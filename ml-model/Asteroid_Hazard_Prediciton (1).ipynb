# Asteroid Hazard Prediciton using Machine Learning  

This project uses asteroids data dervied from NASA's NEO API.
The goal is to predict whether an asteroid is hazardous using machine learning.

## Problem Statement
Given diameter , velocity , distance from Earth of asteroids , how can we use it to predict if asteroid is hazardous or not.

## Approach
### Data
Data from NASA NEO dataset (kaggle)
### Logistic Regression
Used for the process of classification of asteroids into hazardous or not.
### Model Evaluation
##### Accuracy
##### Classification Report
##### Confusion Matrix
## Key Challenge
The dataset was highly imbalanced (90% non-hazardous , 10% hazardous) , leading to misleading accuracy.
## Solution
Applied class weight balancing to improve detection of hazardous asteroids.
## Result
Hazardous asteroid prediction improved to 77%.

```python
import pandas as pd
df = pd.read_csv('neo.csv')
df.head()
```

```python
X = df.drop(columns=['name' , 'id' ,'sentry_object' , 'orbiting_body'  , 'absolute_magnitude' , 'hazardous'])
```

```python
y= df['hazardous']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000 , class_weight="balanced")
model.fit(X_train , y_train)
y_predict= model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report
print(accuracy_score(y_test , y_predict))
print(classification_report(y_test , y_predict))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm=confusion_matrix(y_test,y_predict)

plt.figure(figsize=(5,4))
sns.heatmap(cm , annot=True ,fmt='d' , cmap='Blues')
plt.xlabel("Predictions")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
